## Plotting the Keeling curve
Monthly mean carbon dioxide has been measured at Mauna Loa Observatory, Hawaii, since 1958.  The carbon dioxide data on Mauna Loa constitute the longest record of direct measurements of CO$_2$ in the atmosphere.

This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
# Addins for data reading and plotting
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the workbook and where to find data
spreadsheet_name = 'co2_mm_mlo.xlsx'
worksheet_name = 'co2_mm_mlo'
df = pd.read_excel(spreadsheet_name, sheet_name=worksheet_name)
```

```python
#showing the dataframe
df.head()
```

```python
#Generating the plot
fig, ax = plt.subplots(figsize=(10, 5)) #size of plot
ax.plot(df['year'],df['average'],'r') #average CO2 per year
ax.plot(df['year'],df['deseasonalized'],'k') #deseasonalised CO2 per year
ax.set_xlabel('Year') #text for the x-axis label
ax.set_ylabel('Average Mauna Loa Monthly CO2 (ppm)') #text for the y-axis label
ax.set_title('Monthly Mean CO2 at Mauna Loa Observatory') #text for the title
legend = ax.legend(['average','deseasonalized']) #text for the legend
ax.set_xlim([1950,2030]) #interval for the x-axis
ax.set_ylim([300,450]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
```

```python
#Generating plot area of interest - 1970 to 1980
fig, ax = plt.subplots(figsize=(10, 5)) #size of plot
ax.plot(df['year'],df['average'],'r') #average CO2 per year
ax.plot(df['year'],df['deseasonalized'],'k') #deseasonalised CO2 per year
ax.set_xlabel('Year') #text for the x-axis label
ax.set_ylabel('Average Mauna Loa Monthly CO2 (ppm)') #text for the y-axis label
ax.set_title('Monthly Mean CO2 at Mauna Loa Observatory') #text for the title
legend = ax.legend(['average','deseasonalized'], loc = 'upper left') #text for the legend
ax.set_xlim([1970,1980]) #interval for the x-axis
ax.set_ylim([300,350]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
```
